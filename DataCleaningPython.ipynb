# Data Cleaning in Pandas

```python
import pandas as pd
```

```python
df=pd.read_excel("/Users/bididudy/Downloads/Python/Customer Call List.xlsx") #reading excel file into df
```

```python
df
```

```python
# Data is not cleaned.
# last name column there are some white spaces
# mobile number column has wrong data and null values
# address column has null vales and address not complete
# in paying cutomer column some places instead of Yes only Y is typed and same for N for no.
# same problem as paying cutomer in do not contact column.
```

```python
# 1st thing to do is drop duplicates
df=df.drop_duplicates()
```

```python
# search for not useful columns and get rid of them. Here last column is not useful
df=df.drop(columns = 'Not_Useful_Column')
```

```python
df
```

```python
# next search for any column has white spaces or unwanted characters which needs to be removed. This method is to remove step by step.
#df['Last_Name'] = df['Last_Name'].str.lstrip("/")
#df['Last_Name'] = df['Last_Name'].str.lstrip("...")
#df['Last_Name'] = df['Last_Name'].str.rstrip("_")
#df
```

```python
# we cn remove all unwated symols and white spaces at once.
df['Last_Name'] = df['Last_Name'].str.strip("123./_")
df
```

```python
# phone number column has varius formats of numbers. we need to do them in one format
df['Phone_Number'] = df['Phone_Number'].str.replace("[^a-zA-Z0-9]", '') 
# first we removed all the special characters from all the rows
```

```python
# we apply lambda function to add - after 3 digits
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
# gave error becasue our column is mixed of numbers and strings.
```

```python
# we need to convert whole column into string 
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
```

```python
# we apply lambda function to add - after 3 digits
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df
```

```python
# now we need to remove na and Na 
df['Phone_Number'] = df['Phone_Number'].str.strip("nan--")
df['Phone_Number'] = df['Phone_Number'].str.strip("Na--")
df
```

```python
# in the adress column there are 3 values separated by coma. for each value we need to create column
#df['Address'].str.split(',', 2, expand=True) # here we made 3 columns from Address column.
df[["Street_Address", "State", "Zip_Code"]] = df['Address'].str.split(',', 2, expand=True) #creating 3 columns
df
```

```python
# in paying_customer column we have two variables of yes and no. we need to chnage them to Y and N only.
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No','N')
df
```

```python
# in Do_Not_Contact column we have two variables of yes and no. we need to chnage them to Y and N only.
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes','Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No','N')
df
```

```python
# now we need to remove "N/a" from entire database
df = df.replace('N/a','')
```

```python
df
```

```python
# now we need to remove 'NaN'. NaN is basically null value so we need to use fillna function
df=df.fillna('')
```

```python
df
```

```python
# after basic cleaning is done, bfore giving data call centre we need to ceck column do not contact.
# if that says yes we can't give that client innfo to call centre. 
# so we remove those columns where remark is yes

for x in df.index:
    if df.loc[x, 'Do_Not_Contact'] == 'Y':
        df.drop(x, inplace=True)
        
df
```

```python
# as we dont know what was there in blanck rows in do not contact column we assume it as No and replace it with N
df['Do_Not_Contact'].str.replace('','N') # first we got rid of from white spaces
#df['Do_Not_Contact'].fillna('N')
```

```python
# Phone number column has many blanck rows. those are not useful. we need to remove them.
for x in df.index:
    if df.loc[x, 'Phone_Number'] == '':
        df.drop(x, inplace=True)
        
df
```
